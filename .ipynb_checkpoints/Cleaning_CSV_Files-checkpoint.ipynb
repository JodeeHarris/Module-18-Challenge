```python
#Import dependencies

import pandas as pd
import numpy as np
```

```python
#Import CSV Files

Spring2017 = pd.read_csv("Original_CSV_2017/JC-201701-citibike-tripdata.csv")
Summer2017 = pd.read_csv("Original_CSV_2017/JC-201704-citibike-tripdata.csv")
Fall2017 = pd.read_csv("Original_CSV_2017/JC-201708 citibike-tripdata.csv")
Winter2017 = pd.read_csv("Original_CSV_2017/JC-201712-citibike-tripdata.csv")
Spring2022 = pd.read_csv("Original_CSV_2022/JC-202201-citibike-tripdata.csv")
Summer2022 = pd.read_csv("Original_CSV_2022/JC-202204-citibike-tripdata.csv")
Fall2022 = pd.read_csv("Original_CSV_2022/JC-202208-citibike-tripdata.csv")
Winter2022 = pd.read_csv("Original_CSV_2022/JC-202212-citibike-tripdata.csv")
```

```python
#Removing nan values in the Column

Spring2017 = Spring2017[Spring2017['User Type']!=np.nan]
Summer2017 = Summer2017[Summer2017['usertype']!=np.nan]
Fall2017 = Fall2017[Fall2017['usertype']!=np.nan]
Winter2017 = Winter2017[Winter2017['usertype']!=np.nan]
```

```python
#Renaming Categories

Spring2017 = Spring2017.replace({"User Type":{'Subscriber':'member', 'Customer':'casual'}})
Summer2017 = Summer2017.replace({"usertype":{'Subscriber':'member', 'Customer':'casual'}})
Fall2017 = Fall2017.replace({"usertype":{'Subscriber':'member', 'Customer':'casual'}})
Winter2017 = Winter2017.replace({"usertype":{'Subscriber':'member', 'Customer':'casual'}})
```

```python
#listing categories of dataframes

all_csv = list([Spring2017, Summer2017, Fall2017, Winter2017, Spring2022, Summer2022, Fall2022, Winter2022])
for df in all_csv:
    print(df.columns.tolist())
```

```python
#renaming columns to following year's dataframes

Spring2017 = Spring2017.rename(columns={'Trip Duration': 'tripduration', 'Start Time':'starttime', 'Stop Time':'stoptime', 'Start Station ID':'start station id', 'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude', 'Start Station Longitude': 'start station longitude', 'End Station ID':'end station id', 'End Station Name':'end station name', 'End Station Latitude':'end station latitude', 'End Station Longitude':'end station longitude', 'Bike ID': 'bikeid', "User Type" : "usertype", 'Birth Year': 'birth year', 'Gender':'gender'})
```

```python
#Dropping all NA Values

Spring2017 = Spring2017.dropna()
Summer2017 = Summer2017.dropna()
Fall2017 = Fall2017.dropna()
Winter2017 = Winter2017.dropna()
```

```python
#Dropping all NA Values

Spring2022 = Spring2022.dropna()
Summer2022 = Summer2022.dropna()
Fall2022 = Fall2022.dropna()
Winter2022 = Winter2022.dropna()
```

```python
#listing categories of dataframes

all_csv = list([Spring2017, Summer2017, Fall2017, Winter2017, Spring2022, Summer2022, Fall2022, Winter2022])
for df in all_csv:
    print(df.columns.tolist())
```

```python
# Combined 2017 dataframes into one dataframe.

combine_df2017 = pd.concat([Spring2017, Summer2017, Fall2017, Winter2017],axis=0)
combine_df2017
```

```python
#Dropping non-related/inconsistent dataframes to 2022 data

combine_df2017 = combine_df2017.drop(columns=['tripduration','start station id', 'end station id', 'bikeid', 'birth year', 'gender'])
```

```python
#Combining dataframes for 2022 data

combine_df2022 = pd.concat([Spring2022, Summer2022, Fall2022, Winter2022],axis=0) 
combine_df2022
```

```python
#Dropping non-related/inconsistent dataframes to 2017 data
combine_df2022 = combine_df2022.drop(columns=['ride_id', 'rideable_type','start_station_id','end_station_id'])
```

```python
#listing categories of dataframes to check differences in category names

all_csv = list([combine_df2017, combine_df2022])
for df in all_csv:
    print(df.columns.tolist())
```

```python
#final_names =['start_time','end_time', 'station_start_name','station_end_name','station_start','station_end','start_lat', 'start_long', 'end_lat','end_long', 'user_type']
```

```python
#Renaming categories to match each dataframe

combine_df2017 = combine_df2017.rename(columns={'usertype': 'user_type', 'start station name': 'station_start_name', 'starttime':'start_time', 'start station latitude': 'start_lat', 'start station longitude': 'start_long', 'end station name': 'station_end_name', 'stoptime': 'end_time', 'end station latitude': 'end_lat', 'end station longitude': 'end_long'})
```

```python
#Renaming categories to match each dataframe

combine_df2022 = combine_df2022.rename(columns={'member_casual': 'user_type', 'start_station_name': 'station_start_name', 'started_at':'start_time', 'start_lat': 'start_lat', 'start_lng': 'start_long', 'end_station_name': 'station_end_name', 'ended_at': 'end_time', 'end_lat': 'end_lat', 'end_lng': 'end_long'})
```

```python
#listing categories of dataframes to ensure the categories match

all_csv = list([combine_df2017, combine_df2022])
for df in all_csv:
    print(df.columns.tolist())
```

```python
#Combine the dataframes

final_df = pd.concat([combine_df2017, combine_df2022],axis=0) 
final_df
```

```python
import geopy.distance

def get_distance(row):
    
    coords_1 = row[['start_lat', 'start_long']]
    coords_2 = row[['end_lat', 'end_long']]
    
    return geopy.distance.geodesic(coords_1, coords_2).miles

final_df['miles'] = final_df.apply(get_distance, axis=1)

final_df['miles'].describe()
```

```python
final_df["time_travel_seconds"] = (pd.to_datetime(final_df['end_time']) - pd.to_datetime(final_df['start_time'])).dt.seconds
```

```python
final_df_s = final_df.sample(1000)
```

```python
final_df_s["time_travel_seconds"].plot.box()
```
